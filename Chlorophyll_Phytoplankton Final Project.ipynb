```python
import pandas as pd
import matplotlib.pyplot as plt 
import csv
import numpy as np
```

```python
df = pd.read_csv("Annual_chlorophyll_temp_38N_68W.csv")
```

```python
print(df.head())
```

```python
date = df["Year"]
chla = df["Chlorophyll (mg/m3)"]
temp = df["Temperature (C)"]
```

```python
print(date.head())
print(chla.head())
print(temp.head())
```

```python
plt.clf()
plt.plot(date, chla, color = "purple")
plt.ylabel("amount of chlorophyll")
plt.legend(loc = 4)
plt.twinx()
plt.plot(date, temp)
plt.xlabel("temperature of the water")
plt.title("Measurement of chlorophyll and temperature over a period of time")
plt.legend(loc = 0)
plt.show()
```
